 # RP - GPFA: Textured Bouncing Ball
 
 Recognition Parametrised Gaussian Process Factor Analysis on the textured Bouncing Ball Experiment.
 
 Inference uses a **Variational Bound** on the Free Energy.

```python
import torch
import numpy as np
from matplotlib import pyplot as plt
from utils_process import linear_regression_1D_latent
from utils import generate_2D_latent, generate_skewed_pixel_from_latent, diagonalize

from unstructured_recognition_gpfa import UnstructuredRecognition
```

## Generate Data

```python
# data type
data_type = torch.float32
torch.set_default_dtype(data_type)

# GPUs ?
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
if torch.cuda.is_available():
    print('GP-RPM on GPU')
else:
    print('GP-RPM on CPU')

# Reproducibility
np.random.seed(1)
torch.manual_seed(1)

# Dimension of the observations
num_observation = 50
dim_observation = 10
len_observation = 50
num_inducing = 20


# Generative Model Params
xp_name = 'standard'
scale_th = 0.15
shape_max_0 = 1000
sigma2 = 0.01# Dimension of the observations
num_observation = 50
dim_observation = 10
len_observation = 50
num_inducing = 20


# Generative Model Params
xp_name = 'standard'
scale_th = 0.15
shape_max_0 = 1000
sigma2 = 0.01

# Sampling Frequency
F = 10

# Length of Each sample [sec]
T = int(len_observation / F)

# Sampling Frequency
F = 10

# Length of Each sample [sec]
T = int(len_observation / F)

# Oscillation Frequency
omega = 0.5

# Random initializations
theta = 2*np.pi*np.random.rand(num_observation)
z0 = torch.tensor(np.array([np.cos(theta), np.sin(theta)]).T)
zt, _ = generate_2D_latent(T, F, omega, z0)

# True Latent
true_latent_ext = zt[:, 1:, 0] .unsqueeze(-1)

# Sample Observations
observation_samples = generate_skewed_pixel_from_latent(true_latent_ext, dim_observation,
                                                    scale_th=scale_th, sigma2=sigma2, shape_max_0=shape_max_0)
# Move Observations to GPU
observations = (torch.tensor(observation_samples, dtype=data_type, device=device), )

# Set up observation and inducing points locations
observation_locations = torch.arange(0, len_observation, 1, dtype=data_type, device=device).unsqueeze(-1)
inducing_locations = torch.arange(0, len_observation, len_observation / num_inducing, dtype=data_type, device=device)\
                          .unsqueeze(-1)
```

## Plot Observations

```python
# Sample From Generative distribution
Nsample = 5000
meanss = torch.tensor([-1, 1, - 0.64]).unsqueeze(0).repeat(Nsample, 1)
samples = generate_skewed_pixel_from_latent(meanss.unsqueeze(-1), 10, scale_th=0.15, sigma2=0.01, shape_max_0=1000)[:, :, 0]
plot_id = [1, 2]
plot_legend = ['far', 'close']
plot_color = ['k', [0.5, 0.5, 0.5]]

# Select which observation to plot
plot_index = [0, num_observation-1]

plt.figure(figsize=(5 * (1 + len(plot_index)), 3 * 1))

# Plot Generative Distribution
plt.subplot(1, len(plot_index) + 1 , 1)
for nn in range(len(plot_id)):
    samples_cur = samples[:, plot_id[nn]]
    plt.hist(samples_cur, bins=500, density=True, alpha = 0.75, label=plot_legend[nn], color=plot_color[nn])
    plt.hist(samples_cur, bins=500, density=True, alpha=0.75, color='k', histtype='step')
plt.legend(title='Ball-Pixel')
plt.title('Genrative Distribution')

# Plot Observations
for egobs_id in range(len(plot_index)):
    egobs = plot_index[egobs_id]

    plt.subplot(1, len(plot_index) + 1, egobs_id + 2)
    plt.imshow(observation_samples[egobs].transpose(-1, -2), aspect='auto', cmap='gray',
               extent=[0, 1, -1, 1], vmin=-1, vmax = 1)
    plt.title('Observation ' + str(egobs+1) + '/' + str(num_observation))
    plt.ylabel('Pixel Intensity')
    plt.yticks([])
```

## Fit RP-GPFA

```python
fit_params = {'dim_latent': 1,
               'inference_mode': 'VariationalBound',
               'ite_out': 400,
               'optimizer_prior': {'name': 'Adam', 'param': {'lr': 0.5e-3}},
               'optimizer_factors': {'name': 'Adam', 'param': {'lr': 1e-3}},
               'optimizer_inducing_points': {'name': 'Adam', 'param': {'lr': 1e-3}},
               'dim_hidden': ([50, 50],),
               'constraint_factors':'diag'
                }

model = UnstructuredRecognition(observations, observation_locations,
                                inducing_locations=inducing_locations, fit_params=fit_params)
loss_tot = model.fit(observations)
```

## Plot Loss

```python
plt.figure()
plot_full_loss = False
offset = 100
step = model.fit_params['ite_prior'] + model.fit_params['ite_factors'] + model.fit_params['ite_inducing_points']
if model.fit_params['inference_mode'] == 'VariationalBound':
    step += model.fit_params['ite_factors_tilde']

plt.plot(model.loss_tot[offset:-1:step], c='k', linewidth=2)
plt.xlabel('Iterations')
plt.title('- Free Energy')
```

## Plot Model

Use a linear regression to align fit and true latent

```python
# Grasp Variational Marginals
q = model.variational_marginals.suff_stat_mean
current_mean = q[0].detach().clone()
current_vari = diagonalize(q[1].detach().clone().diagonal(dim1=-1, dim2=-2) - current_mean ** 2) 

if torch.cuda.is_available():
    current_mean = current_mean.cpu()
    current_vari = current_vari.cpu()
    current_obs  = observations[0].cpu()

# Linear Regression
latent_true, current_mean, current_vari, _ = linear_regression_1D_latent(true_latent_ext, current_mean, current_vari)
```

Plot Observation and true latent

```python
# Obs to be plotted
plot_indices = [0, num_observation -1]

# Problem dimensions
num_observation, len_observation, dim_observation = observations[0].shape
dim_latent = 1

plt.figure(figsize=(5 * 3, 3 * (1+len(plot_indices))))
for nn_id in range(len(plot_indices)):

    nn = plot_indices[nn_id]

    # Plot All Pixels in 2D
    plt.subplot(len(plot_indices), 2,  1 + nn_id)
    plt.imshow(current_obs[nn].transpose(-1, -2),
               aspect='auto', cmap='gray', vmin=-1, vmax=1, extent=[0, 1, -1, 1], origin='lower')
    plt.autoscale(enable=True, axis='x', tight=True)
    #plt.ylabel('n =  ' + str(nn) + '/' + str(num_observation))
    plt.grid()
    plt.yticks([])
    plt.title('Observation ' + str(nn+1) + '/' + str(num_observation))


    # PLot fitted latents
    plt.subplot(len(plot_indices), 2,  1 + nn_id + len(plot_indices))
    cmap = plt.get_cmap("tab10")
    for dim_latent_cur in range(dim_latent):

        true = latent_true[nn].squeeze()
        xx = np.arange(len(true)) / len(true)
        plt.plot(xx, true, color='k', label='True', linestyle='-.')

        
        mm = current_mean[nn].squeeze()
        vv = 2 * np.sqrt(current_vari[nn]).squeeze()
 
        plt.plot(xx, mm, color='tab:blue', label='RP-GPFA', linewidth=2)
        plt.fill_between(xx, mm - vv, mm + vv, color='tab:blue', alpha=.1)

    plt.autoscale(enable=True, axis='x', tight=True)
    plt.ylim([-1.2, 1.2])
    plt.yticks([-1, 0, 1])
    plt.grid()
    if nn_id == 0:
        plt.legend(loc=1, title='Infered Latent')
    plt.xlabel('Time [a.u]')
    plt.xticks([0, 0.5, 1])
```
